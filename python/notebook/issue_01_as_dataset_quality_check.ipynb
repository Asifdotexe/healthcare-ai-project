```python
import os
import kagglehub
import pandas as pd
```

```python
# why use kagglehub? 
# better for pipelines where we need updated or want to update data periodically

# diabetes dataset with age, gender, blood-type, medical-condition, medication
path = kagglehub.dataset_download("prasad22/healthcare-dataset")
diabetes_prasad = pd.read_csv(os.path.join(path,'healthcare_dataset.csv'))

# diabetes dataset with id, pregnancies, glucose, bloodPressure, skinThickness, Insulin, BMI, DiabetesPedigreeFunction, Age
# potential issue is that this is just for females
path = kagglehub.dataset_download("nanditapore/healthcare-diabetes")
diabetes_nanditapore = pd.read_csv(os.path.join(path,'Healthcare-Diabetes.csv'))

# this is a wider dataset with a lot of different normalized file containing symptoms, precaution and severity information
path = kagglehub.dataset_download("itachi9604/disease-symptom-description-dataset")
symptoms_master = pd.read_csv(os.path.join(path,'dataset.csv'))
symptoms_severity = pd.read_csv(os.path.join(path,'Symptom-severity.csv'))
symptoms_description = pd.read_csv(os.path.join(path,'symptom_Description.csv'))
symptoms_precaution = pd.read_csv(os.path.join(path,'symptom_precaution.csv'))

# this is a simpler dataset that works on the same principles as the one above, deals with symptom to disease classification.
path = kagglehub.dataset_download("uom190346a/disease-symptoms-and-patient-profile-dataset")
symptom_to_disease = pd.read_csv(os.path.join(path,'Disease_symptom_and_patient_profile_dataset.csv'))

# this dataset contains 11,000 medicine details with their names, dosages, and side effects.
path = kagglehub.dataset_download("singhnavjot2062001/11000-medicine-details")
eleven_k_meds = pd.read_csv(os.path.join(path,'Medicine_Details.csv'))
```
